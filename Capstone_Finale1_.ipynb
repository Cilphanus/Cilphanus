```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
df1 = pd.read_csv('/content/Life-Expectancy-Data-Updated.csv')
df2 = pd.read_csv('/content/diabetes.csv')

# Dataset 1 - Health Indicators
# Display basic information about the dataset
print("Dataset 1 Info:\n", df1.info())

# Check for missing values
missing_values_df1 = df1.isnull().sum()
print("\nMissing Values in Dataset 1:\n", missing_values_df1)

# Handle missing values
df1.fillna(df1.median(), inplace=True)

# Visualize the correlation matrix for Dataset 1
correlation_matrix_df1 = df1.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix_df1, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix - Dataset 1')
plt.show()

# Dataset 2 - Diabetes Dataset
# Display basic information about the dataset
print("\nDataset 2 Info:\n", df2.info())

# Check for missing values
missing_values_df2 = df2.isnull().sum()
print("\nMissing Values in Dataset 2:\n", missing_values_df2)

# Visualize the correlation matrix for Dataset 2
correlation_matrix_df2 = df2.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix_df2, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix - Dataset 2')
plt.show()

# Integration
# Assuming there's a common column between the two datasets for integration, e.g., 'Country' and 'Region'
merged_df = pd.merge(df1, df2, how='inner', left_on=['BMI', 'BMI'], right_on=['BMI', 'BMI'])

# Display basic information about the merged dataset
print("\nMerged Dataset Info:\n", merged_df.info())

# Visualize the correlation matrix for the merged dataset
correlation_matrix_merged = merged_df.corr()
plt.figure(figsize=(15, 12))
sns.heatmap(correlation_matrix_merged, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix - Merged Dataset')
plt.show()
```
