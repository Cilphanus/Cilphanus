```python

import matplotlib.pyplot as plt

x = [10, 24, 33, 46, 58, 62, 70, 88, 91]
y1 = [11, 32, 52, 37, 16, 32, 54, 36, 18]
y2 = [20, 44, 63, 45, 27, 48, 69, 48, 27]

plt.plot(x, y1, label="Number of customers today",color="#131720")
plt.plot(x, y2, label="Number of customers per month",color="r")
plt.plot()

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Customers Attendance")
plt.legend()
plt.show()
```
